```python
import pandas as pd
import os
import plotly.graph_objects as go

data_dir = "/kaggle/input/bank-account-fraud-dataset-neurips-2022/"
file_names = ["Base.csv", "Variant I.csv", "Variant II.csv", "Variant III.csv", "Variant IV.csv", "Variant V.csv"]

datasets = {file_name: pd.read_csv(os.path.join(data_dir, file_name)) for file_name in file_names}

summary_data = []

for name, df in datasets.items():
    row_count = df.shape[0]
    col_count = df.shape[1]

    #check column
    if 'fraud_bool' in df.columns:
        fraud_count = df[df['fraud_bool'] == 1].shape[0]  
        fraud_ratio = (fraud_count / row_count) * 100
    else:
        fraud_count = "N/A"
        fraud_ratio = "N/A"

    summary_data.append([name, row_count, col_count, fraud_count, fraud_ratio])

summary_df = pd.DataFrame(summary_data, columns=["Dataset", "Rows", "Columns", "Fraud Cases", "Fraud %"])

#plot
fig = go.Figure(data=[go.Table(
    header=dict(values=list(summary_df.columns), fill_color='lightgray', align='left'),
    cells=dict(values=[summary_df[col] for col in summary_df.columns], align='left')
)])

fig.update_layout(title="Comparison of Bank Account Fraud Datasets")
fig.show()
```

```python
import numpy as np
import pandas as pd


engineered_datasets = {}

for name, df in datasets.items():
    df = df.copy()  

    #avoid division by zero by replacing 0s with a small value
    #i might need to change this for ML or transformer models later on)
    epsilon = 1e-6  

    #transaction behavior
    if "velocity_24h" in df.columns:
        df["velocity_1h"] = df["velocity_24h"] / 24  
    if "current_address_months_count" in df.columns and "customer_age" in df.columns:
        df["account_age_ratio"] = df["current_address_months_count"] / (df["customer_age"] + epsilon)
    if "intended_balcon_amount" in df.columns and "income" in df.columns:
        df["spending_ratio"] = df["intended_balcon_amount"] / (df["income"] + epsilon)

    #risk 
    if "device_distinct_emails_8w" in df.columns and "email_is_free" in df.columns:
        df["email_risk"] = df["device_distinct_emails_8w"] / (df["email_is_free"] + epsilon)
    if "phone_home_valid" in df.columns and "phone_mobile_valid" in df.columns:
        df["phone_risk"] = df["phone_home_valid"] / (df["phone_mobile_valid"] + epsilon)

    #historical patterns
    if "session_length_in_minutes" in df.columns:
        df["session_anomaly"] = df["session_length_in_minutes"] / (df["session_length_in_minutes"].mean() + epsilon)
    if "device_fraud_count" in df.columns and "device_distinct_emails_8w" in df.columns:
        df["device_risk"] = df["device_fraud_count"] / (df["device_distinct_emails_8w"] + epsilon)
    if "zip_count_4w" in df.columns:
        df["zip_risk"] = df["zip_count_4w"] / (df["zip_count_4w"].mean() + epsilon)

    #interaction
    if "intended_balcon_amount" in df.columns and "proposed_credit_limit" in df.columns:
        df["credit_utilization"] = df["intended_balcon_amount"] / (df["proposed_credit_limit"] + epsilon)
    if "bank_months_count" in df.columns and "bank_branch_count" in df.columns:
        df["bank_activity_ratio"] = df["bank_months_count"] / (df["bank_branch_count"] + epsilon)

    #normalize
    skewed_features = ["spending_ratio", "session_anomaly", "credit_utilization"]
    for feature in skewed_features:
        if feature in df.columns:
            df[feature] = np.where(df[feature] < 0, 0, df[feature]) 
            df[feature] = np.log1p(df[feature])  

    engineered_datasets[name] = df

print("âœ… Feature engineering complete! Engineered features added.")
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, precision_recall_curve, roc_curve, auc
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
from IPython.display import display


df = engineered_datasets["Base.csv"].copy()

#features
y = df["fraud_bool"]
X = pd.get_dummies(df.drop(columns=["fraud_bool"]), drop_first=True)
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# split 
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.15, stratify=y_train_full, random_state=42)


fraud_ratio = y_train.value_counts(normalize=True)[1]
scale_pos_weight = round((1 - fraud_ratio) / fraud_ratio, 2)
print(f"\nðŸ“Š Fraud Ratio: {fraud_ratio:.4f}, scale_pos_weight: {scale_pos_weight}")

#models
models = {
    "XGBoost (GPU)": xgb.XGBClassifier(
        tree_method="hist", device="cuda", use_label_encoder=False,
        eval_metric="aucpr", learning_rate=0.03, max_depth=6,
        n_estimators=1000, subsample=0.85, colsample_bytree=0.85,
        scale_pos_weight=scale_pos_weight, early_stopping_rounds=25, random_state=42
    ),
    "LightGBM (GPU)": lgb.LGBMClassifier(
        device="gpu", boosting_type="gbdt", n_estimators=1000, learning_rate=0.03,
        max_depth=6, subsample=0.85, colsample_bytree=0.85, is_unbalance=True, random_state=42
    ),
    "CatBoost (GPU)": CatBoostClassifier(
        task_type="GPU", iterations=1000, learning_rate=0.03, depth=6,
        random_seed=42, verbose=0, early_stopping_rounds=25,
        class_weights=[1.0, scale_pos_weight]
    )
}

#train
results = []
probs_dict = {}

for name, model in models.items():
    print(f"\nðŸš€ Training {name}...")
    if "XGBoost" in name:
        model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=100)
    elif "CatBoost" in name:
        model.fit(X_train, y_train, eval_set=(X_val, y_val))
    elif "LightGBM" in name:
        model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric="auc")

    #probs for post-eval
    probs_dict[name] = model.predict_proba(X_test)[:, 1]
    y_pred_default = model.predict(X_test)
    report = classification_report(y_test, y_pred_default, output_dict=True)
    auc_score = roc_auc_score(y_test, probs_dict[name])

    results.append({
        "Model": name,
        "Test Precision": round(report["1"]["precision"], 2),
        "Test Recall": round(report["1"]["recall"], 2),
        "Test F1": round(report["1"]["f1-score"], 2),
        "Test AUC": round(auc_score, 4),
        "Confusion Matrix": confusion_matrix(y_test, y_pred_default).tolist()
    })

summary_df = pd.DataFrame(results)
display(summary_df)

#threshold tuning
best_threshold_results = []

for name, probs in probs_dict.items():
    precision, recall, thresholds = precision_recall_curve(y_test, probs)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-6)
    best_idx = np.argmax(f1_scores)
    best_thresh = thresholds[best_idx]
    best_pred = (probs > best_thresh).astype(int)
    report = classification_report(y_test, best_pred, output_dict=True)

    best_threshold_results.append({
        "Model": name,
        "Best Threshold": round(best_thresh, 4),
        "Best Precision": round(report["1"]["precision"], 2),
        "Best Recall": round(report["1"]["recall"], 2),
        "Best F1": round(report["1"]["f1-score"], 2)
    })

print("\n Best F1-Score Thresholds:")
display(pd.DataFrame(best_threshold_results))

#plot pr and roc
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
for name, probs in probs_dict.items():
    precision, recall, _ = precision_recall_curve(y_test, probs)
    plt.plot(recall, precision, label=name)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title("Precision-Recall Curve"); plt.legend()

plt.subplot(1, 2, 2)
for name, probs in probs_dict.items():
    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.4f})")
plt.plot([0, 1], [0, 1], "k--")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curve"); plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, roc_auc_score, roc_curve, 
                             confusion_matrix, precision_recall_curve, make_scorer)
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt


df = engineered_datasets["Base.csv"].copy()

y = df["fraud_bool"]
X = pd.get_dummies(df.drop(columns=["fraud_bool"]), drop_first=True)

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)


X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_scaled, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.15, stratify=y_train_full, random_state=42
)


num_negative = (y_train_full == 0).sum()
num_positive = (y_train_full == 1).sum()
pos_weight_value = num_negative / num_positive
print(f"info original pos_weight {pos_weight_value:.4f}")


smote = SMOTE(sampling_strategy=0.3, random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")



#cost sensitv

class FocalLoss(nn.Module):
    def __init__(self, alpha=0.25, gamma=2.0, pos_weight=None):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.pos_weight = pos_weight

    def forward(self, inputs, targets):
        
        bce_loss = F.binary_cross_entropy_with_logits(
            inputs, targets, reduction='none', pos_weight=self.pos_weight
        )
        pt = torch.exp(-bce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * bce_loss
        return focal_loss.mean()



#cnn

class FraudNet(nn.Module):
    def __init__(self, input_dim, hidden1=256, hidden2=128, hidden3=64, dropout=0.3):
        super(FraudNet, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden1)
        self.bn1 = nn.BatchNorm1d(hidden1)
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.bn2 = nn.BatchNorm1d(hidden2)
        self.fc3 = nn.Linear(hidden2, hidden3)
        self.bn3 = nn.BatchNorm1d(hidden3)
        self.fc4 = nn.Linear(hidden3, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
       
        residual = x.mean(dim=1, keepdim=True)
        x = F.relu(self.bn1(self.fc1(x)))
        x = self.dropout(x)
        x = F.relu(self.bn2(self.fc2(x)))
        x = self.dropout(x)
        x = F.relu(self.bn3(self.fc3(x)))
        x = self.dropout(x)
        x = self.fc4(x) + residual
        return torch.sigmoid(x)


#sklearn

class SklearnCNN(BaseEstimator, ClassifierMixin):
    def __init__(self, 
                 input_dim,
                 epochs=10, 
                 lr=0.001, 
                 alpha=0.25, 
                 gamma=2.0, 
                 pos_weight=None,
                 hidden1=256,
                 hidden2=128,
                 hidden3=64,
                 dropout=0.3):

        self.input_dim = input_dim
        self.epochs = epochs
        self.lr = lr
        self.alpha = alpha
        self.gamma = gamma
        self.pos_weight = pos_weight
        self.hidden1 = hidden1
        self.hidden2 = hidden2
        self.hidden3 = hidden3
        self.dropout = dropout

        self.model = FraudNet(input_dim, hidden1, hidden2, hidden3, dropout).to(device)
        self.criterion = FocalLoss(alpha=self.alpha, gamma=self.gamma, pos_weight=self.pos_weight)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.lr)
        #  scheduler to reduce lr when training
        self.scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer, mode='min', patience=2, factor=0.5, verbose=False
        )
        self.best_threshold = 0.5

    def fit(self, X, y):
        
        X_array = X.values if hasattr(X, 'values') else X
        y_array = y.values if hasattr(y, 'values') else y
        X_tensor = torch.tensor(X_array, dtype=torch.float32).to(device)
        y_tensor = torch.tensor(y_array, dtype=torch.float32).view(-1, 1).to(device)

        
        global tensor_X_val, tensor_y_val

        for epoch in range(self.epochs):
            
            self.model.train()
            self.optimizer.zero_grad()
            outputs = self.model(X_tensor)
            loss = self.criterion(outputs, y_tensor)
            loss.backward()
            self.optimizer.step()
            self.scheduler.step(loss.item())

            # auto threshold tuning for val set
            self.model.eval()
            with torch.no_grad():
                val_probs = self.model(tensor_X_val).cpu().numpy().flatten()
                precision, recall, thresholds = precision_recall_curve(y_val, val_probs)
                f1_scores = 2 * (precision * recall) / (precision + recall + 1e-6)
                best_idx = np.argmax(f1_scores)
                if thresholds.size > 0:
                    self.best_threshold = thresholds[best_idx]

            print(f"epoch {epoch+1:02d} loss {loss.item():.4f} best threshold {self.best_threshold:.4f}")

        return self

    def predict_proba(self, X):
        X_array = X.values if hasattr(X, 'values') else X
        X_tensor = torch.tensor(X_array, dtype=torch.float32).to(device)
        self.model.eval()
        with torch.no_grad():
            probs = self.model(X_tensor).cpu().numpy().flatten()
        return np.vstack([1 - probs, probs]).T

    def predict(self, X):
        probs = self.predict_proba(X)[:, 1]
        return (probs > self.best_threshold).astype(int)

    
    def get_params(self, deep=True):
        return {
            'input_dim': self.input_dim,
            'epochs': self.epochs,
            'lr': self.lr,
            'alpha': self.alpha,
            'gamma': self.gamma,
            'pos_weight': self.pos_weight,
            'hidden1': self.hidden1,
            'hidden2': self.hidden2,
            'hidden3': self.hidden3,
            'dropout': self.dropout
        }

    def set_params(self, **params):
        for key, value in params.items():
            setattr(self, key, value)
       
        self.model = FraudNet(
            self.input_dim, self.hidden1, self.hidden2, self.hidden3, self.dropout
        ).to(device)
        self.criterion = FocalLoss(
            alpha=self.alpha, gamma=self.gamma, pos_weight=self.pos_weight
        )
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.lr)
        self.scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer, mode='min', patience=2, factor=0.5, verbose=False
        )
        return self


#cost scorer

def cost_based_metric(y_true, y_pred):

    cost_fn = 5.0
    cost_fp = 1.0
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    total_cost = (cost_fn * fn) + (cost_fp * fp)
    return -total_cost

cost_scorer = make_scorer(cost_based_metric, greater_is_better=True)



#hyper search

param_distributions = {
    'epochs': [5, 10, 15],
    'lr': [1e-3, 5e-4, 1e-4],
    'alpha': [0.25, 0.5],
    'gamma': [2.0, 3.0],
    'hidden1': [256, 512],
    'hidden2': [128, 256],
    'hidden3': [64, 128],
    'dropout': [0.3, 0.5]
}

#val thresh tunning
tensor_X_val = torch.tensor(X_val.values, dtype=torch.float32).to(device)
tensor_y_val = torch.tensor(y_val.values, dtype=torch.float32).view(-1,1).to(device)

base_cnn = SklearnCNN(
    input_dim=X_train_res.shape[1],
    epochs=5,
    lr=1e-3,
    alpha=0.25,
    gamma=2.0,
    pos_weight=torch.tensor([scale_pos_weight]).to(device)
)

#random search
search = RandomizedSearchCV(
    estimator=base_cnn,
    param_distributions=param_distributions,
    n_iter=5,  
    scoring=cost_scorer,
    cv=3,
    verbose=1,
    random_state=42
)

print("\ninfo starting randomizedsearchcv")
search.fit(X_train_res, y_train_res)
print("info finished randomizedsearchcv")

best_cnn = search.best_estimator_
print(f"info best parameters found {search.best_params_}")



#final train


y_pred = best_cnn.predict(X_test)
y_pred_proba = best_cnn.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("\ncnn classification report test set")
print(classification_report(y_test, y_pred))
print(f"cnn roc auc score {roc_auc:.4f}")

cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print("\nconfusion matrix")
print(cm)
print("\ndetection details")
print(f"frauds correctly detected tp {tp}")
print(f"frauds missed fn {fn}")
print(f"nonfrauds flagged as fraud fp {fp}")

#roc plot
plt.figure(figsize=(10, 5))
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label="cnn")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.title("roc curve")
plt.legend()
plt.show()
```
