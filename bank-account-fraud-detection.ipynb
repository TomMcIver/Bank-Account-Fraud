```python
import pandas as pd
import os
import plotly.graph_objects as go


data_dir = "/kaggle/input/bank-account-fraud-dataset-neurips-2022/"
file_names = ["Base.csv", "Variant I.csv", "Variant II.csv", "Variant III.csv", "Variant IV.csv", "Variant V.csv"]

#load data
datasets = {file_name: pd.read_csv(os.path.join(data_dir, file_name)) for file_name in file_names}


summary_data = []

for name, df in datasets.items():
    row_count = df.shape[0]
    col_count = df.shape[1]
    
    # check if fraud cplumn is there
    if 'fraud_bool' in df.columns:
        fraud_count = df[df['fraud_bool'] == 1].shape[0]  
        fraud_ratio = (fraud_count / row_count) * 100
    else:
        fraud_count = "N/A"
        fraud_ratio = "N/A"

    summary_data.append([name, row_count, col_count, fraud_count, fraud_ratio])


summary_df = pd.DataFrame(summary_data, columns=["Dataset", "Rows", "Columns", "Fraud Cases", "Fraud %"])

# plot
fig = go.Figure(data=[go.Table(
    header=dict(values=list(summary_df.columns), fill_color='lightgray', align='left'),
    cells=dict(values=[summary_df[col] for col in summary_df.columns], align='left')
)])

fig.update_layout(title="Comparison of Bank Account Fraud Datasets")
fig.show()
```

```python
import numpy as np

#check missing values
missing_values = {}

for name, df in datasets.items():
    missing_count = df.isnull().sum().sum()
    missing_values[name] = missing_count


missing_df = pd.DataFrame(list(missing_values.items()), columns=["Dataset", "Missing Values"])

#show missing vluse in table 
fig = go.Figure(data=[go.Table(
    header=dict(values=list(missing_df.columns), fill_color='lightgray', align='left'),
    cells=dict(values=[missing_df[col] for col in missing_df.columns], align='left')
)])

fig.update_layout(title="Missing Values in Bank Fraud Datasets")
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from tqdm.notebook import tqdm
import time


selected_features = ["income", "customer_age", "intended_balcon_amount", "session_length_in_minutes"]


plt.figure(figsize=(15, 10))

#eta
total_iterations = len(selected_features) * len(datasets)
progress_bar = tqdm(total=total_iterations, desc="Processing features")

#plot
for i, feature in enumerate(selected_features, 1):
    plt.subplot(2, 2, i)
    
    for name, df in datasets.items():
        if feature in df.columns:
            data = df[feature].values
            finite_mask = np.isfinite(data)
            finite_data = data[finite_mask]
            
            if len(finite_data) > 0:
                x_min, x_max = finite_data.min(), finite_data.max()
                x = np.linspace(x_min, x_max, 1000)

                kde = stats.gaussian_kde(finite_data)
                y = kde(x)

                plt.plot(x, y, label=name, linewidth=2)
        

        progress_bar.update(1)
    
    plt.title(f"Distribution of {feature}")
    plt.legend()

progress_bar.close()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig, axes = plt.subplots(2, 3, figsize=(18, 12))

for i, (name, df) in enumerate(datasets.items()):
    ax = axes[i // 3, i % 3]  # Arrange in a grid

    #get only numeric columns
    df_numeric = df.select_dtypes(include=['number']).replace([np.inf, -np.inf], np.nan)

    #delete columns that are NaN
    df_numeric = df_numeric.dropna(axis=1, how='all')

    #correlation matrix
    if not df_numeric.empty and df_numeric.shape[1] > 1: 
        corr_matrix = df_numeric.corr()

        #  na with 0
        corr_matrix = corr_matrix.fillna(0)

        #plot
        im = ax.imshow(corr_matrix, cmap="coolwarm", vmin=-1, vmax=1)
        ax.set_title(f"Correlation Matrix - {name}")

        if i == 0:
            fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    else:
        ax.set_title(f"No Valid Data - {name}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from tqdm.notebook import tqdm


df = datasets["Base.csv"]

#fraud vs non-fraud cases
fraud_cases = df[df["fraud_bool"] == 1]
non_fraud_cases = df[df["fraud_bool"] == 0]


features = ["income", "customer_age", "intended_balcon_amount", "session_length_in_minutes"]


plt.figure(figsize=(15, 10))
progress_bar = tqdm(total=len(features), desc="Processing Fraud vs. Non-Fraud")

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    
    if feature in df.columns:
        # Handle na
        fraud_data = fraud_cases[feature].replace([np.inf, -np.inf], np.nan).dropna()
        non_fraud_data = non_fraud_cases[feature].replace([np.inf, -np.inf], np.nan).dropna()
        
        if len(fraud_data) > 0 and len(non_fraud_data) > 0:
            x_min = min(fraud_data.min(), non_fraud_data.min())
            x_max = max(fraud_data.max(), non_fraud_data.max())
            x = np.linspace(x_min, x_max, 1000)
            
            #gaussian kde
            kde_fraud = stats.gaussian_kde(fraud_data)
            kde_non_fraud = stats.gaussian_kde(non_fraud_data)
            
            # plot kde 
            plt.plot(x, kde_fraud(x), label="Fraud", linewidth=2, color="red")
            plt.plot(x, kde_non_fraud(x), label="Non-Fraud", linewidth=2, color="blue")
        
    plt.title(f"Fraud vs. Non-Fraud - {feature}")
    plt.legend()
    
    progress_bar.update(1)

progress_bar.close()
plt.tight_layout()
plt.show()
```
